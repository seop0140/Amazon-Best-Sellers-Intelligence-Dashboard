### Exploratory Data Analysis

After cleaning the data, I decided to do an Exploratory Data Analysis(EDA) to understand the dataset before moving into deeper analysis. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('amazon_dataset_pandas.csv', encoding='latin1')
df.head(10)
```

```python
df_product = pd.read_csv('products_table.csv')
df_product.head(10)
```

```python
df_variants = pd.read_csv('variants_table.csv')
df_variants.head(10)
```

Codes above shows the first 10 rows of the amazon_dataset_pandas, products_table and variants_table csv file respectively. 

Now I'm checking whether there are duplicates in the cleaned data:

```python
df_shape = df.shape
df_product_shape = df_product.shape
df_variants_shape = df_variants.shape
print(df_shape, df_product_shape, df_variants_shape)
```

```python
df_shape_dup = df.drop_duplicates().shape
product_shape_dup = df_product.drop_duplicates().shape
variants_shape_dup = df_variants.drop_duplicates().shape
print(df_shape_dup, product_shape_dup, variants_shape_dup)
```

The shape of the dataset after dropping duplicates is the same, so I can assume there are no duplicates. Hence, there is at most one row per columns of each dataset.

### Price distribution

```python

df.sort_values(by = "listedPrice", ascending=False).head(5)
```

As can be seen above, there is an outlier in listedPrice of $23500, which is significantly high compared to other listedPrices. This compressed the majority of product prices into a small range. To better visualize the price distribution, I decided to put the price in log scale.

```python
plt.figure(figsize=(10,5))
plt.hist(np.log1p(df['salePrice'].dropna()), bins=30)
plt.title('Log Distribution of Sale Prices')
plt.xlabel('log(Price)')
plt.ylabel('Number of Products')
plt.show()


# Boxplot for outliers
plt.figure(figsize=(10,2))
plt.boxplot(df['salePrice'].dropna(), vert=False)
plt.title('Sale Price Boxplot')
plt.xlabel('Price')
plt.show()
```

### Rating distribution

```python
plt.figure(figsize=(8,4))
plt.hist(df['rating'].dropna(), bins=5, color='green', edgecolor='black')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Products')
plt.show()
```

### Top 10 product distribution

```python
top_categories = df['product_category'].value_counts().head(10)

plt.figure(figsize=(12,5))
top_categories.plot(kind='bar', color='purple')
plt.title('Top 10 Product Categories')
plt.ylabel('Number of Products')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```

It can be observed that Books category has the most products, and the Health & Household category has the least products.
